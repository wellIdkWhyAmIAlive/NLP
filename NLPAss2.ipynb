```python
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from gensim.models import Word2Vec
!pip install gensim
```

```python
documents = [
    "I love natural language processing",
    "NLP is fun and interesting",
    "I love learning NLP"
]
```

```python
count_vect = CountVectorizer()
bow_counts = count_vect.fit_transform(documents)

print("Vocabulary:", count_vect.get_feature_names_out())
print("Count Matrix:\n", bow_counts.toarray())
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.preprocessing import normalize

count_vect = CountVectorizer()
bow_counts = count_vect.fit_transform(documents)
```

```python
bow_norm = normalize(bow_counts, norm='l1')

print("Normalized Count Matrix:\n", bow_norm.toarray())

```

```python
tfidf_vect = TfidfVectorizer()
tfidf_matrix = tfidf_vect.fit_transform(documents)

print("TF-IDF Vocabulary:", tfidf_vect.get_feature_names_out())
print("TF-IDF Matrix:\n", tfidf_matrix.toarray())
```

```python
tokenized_docs = [doc.lower().split() for doc in documents]

w2v_model = Word2Vec(
    sentences=tokenized_docs,
    vector_size=100,
    window=5,
    min_count=1,
    workers=4
)
```

```python
print("Embedding for 'nlp':\n", w2v_model.wv['nlp'])
```

```python
print("Similar words to 'nlp':", w2v_model.wv.most_similar('nlp'))
```
