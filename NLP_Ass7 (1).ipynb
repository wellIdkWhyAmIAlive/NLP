```python
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.probability import FreqDist

nltk.download('punkt')
nltk.download('stopwords')
nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet')

text = """
Natural Language Processing is a branch of Artificial Intelligence.
It enables computers to understand, analyze, and generate human language.
NLTK is a popular Python library for NLP tasks.
"""

sentences = sent_tokenize(text)
print("Sentence Tokenization:")
for s in sentences:
    print("-", s)

words = word_tokenize(text)
print("\nWord Tokenization:")
print(words)

stop_words = set(stopwords.words('english'))
filtered_words = [w.lower() for w in words if w.isalnum() and w.lower() not in stop_words]
print("\nAfter Stopword Removal:")
print(filtered_words)

pos_tags = nltk.pos_tag(filtered_words)
print("\nPart-of-Speech Tagging:")
print(pos_tags)

stemmer = PorterStemmer()
stemmed_words = [stemmer.stem(w) for w in filtered_words]
print("\nStemming:")
print(stemmed_words)

lemmatizer = WordNetLemmatizer()
lemmatized_words = [lemmatizer.lemmatize(w) for w in filtered_words]
print("\nLemmatization:")
print(lemmatized_words)

freq_dist = FreqDist(filtered_words)
print("\nWord Frequency Distribution:")
for word, freq in freq_dist.most_common():
    print(word, ":", freq)
```
