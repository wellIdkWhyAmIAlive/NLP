```python
from transformers import pipeline

sentiment = pipeline(
    "sentiment-analysis",
    model="nlptown/bert-base-multilingual-uncased-sentiment"
)
```

```python
texts = [
    "यह फिल्म बहुत अच्छी है",
    "मुझे यह प्रोडक्ट पसंद नहीं आया",
    "सेवा ठीक थी लेकिन बहुत अच्छी नहीं",
    "यह अनुभव शानदार था"
]

for text in texts:
    result = sentiment(text)[0]
    print("Text:", text)
    print("Prediction:", result)
    print()
```

```python
def convert_label(star_label):
    stars = int(star_label.split()[0])

    if stars <= 2:
        return "Negative"
    elif stars == 3:
        return "Neutral"
    else:
        return "Positive"

for text in texts:
    result = sentiment(text)[0]
    final_sentiment = convert_label(result['label'])

    print(f"{text} → {final_sentiment}")
```

```python
texts = [
    "Yeh movie bahut acchi hai",
    "Mujhe yeh pasand nahi aaya",
    "Service thik thi"
]

for text in texts:
    result = sentiment(text)[0]
    print(text, "→", result)
```
