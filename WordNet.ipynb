```python
import nltk
from nltk.corpus import wordnet as wn

nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
text = "The happy dog chased the small cat"
words = text.lower().split()
print("Words in text:", words)
```

```python
def get_semantic_relations(word):
    synonyms = set()
    antonyms = set()
    hypernyms = set()

    for syn in wn.synsets(word):

        # Synonyms
        for lemma in syn.lemmas():
            synonyms.add(lemma.name())

            # Antonyms
            if lemma.antonyms():
                antonyms.add(lemma.antonyms()[0].name())

        # Hypernyms
        for hyper in syn.hypernyms():
            for lemma in hyper.lemmas():
                hypernyms.add(lemma.name())

    return synonyms, antonyms, hypernyms
```

```python
for word in words:
    syn, ant, hyp = get_semantic_relations(word)

    print("\nWord:", word)
    print("Synonyms:", syn if syn else "None")
    print("Antonyms:", ant if ant else "None")
    print("Hypernyms:", hyp if hyp else "None")
```
