```python
import nltk
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

import re
import pandas as pd

from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import LabelEncoder
```

```python
data = {
    "text": [
        "I love Natural Language Processing!",
        "Text cleaning is an important step.",
        "TF-IDF converts text into numbers."
    ],
    "label": ["positive", "neutral", "neutral"]
}

df = pd.DataFrame(data)
```

```python
def clean_text(text):
    text = text.lower()
    text = re.sub(r'[^a-z\s]', '', text)
    return text

df["clean_text"] = df["text"].apply(clean_text)
```

```python
stop_words = set(stopwords.words("english"))
lemmatizer = WordNetLemmatizer()

def preprocess(text):
    tokens = word_tokenize(text)
    tokens = [lemmatizer.lemmatize(word) for word in tokens if word not in stop_words]
    return " ".join(tokens)

df["processed_text"] = df["clean_text"].apply(preprocess)
```

```python
encoder = LabelEncoder()
df["label_encoded"] = encoder.fit_transform(df["label"])
```

```python
tfidf = TfidfVectorizer()
tfidf_matrix = tfidf.fit_transform(df["processed_text"])

tfidf_df = pd.DataFrame(
    tfidf_matrix.toarray(),
    columns=tfidf.get_feature_names_out()
)
```

```python
df.to_csv("cleaned_text_output.csv", index=False)
tfidf_df.to_csv("tfidf_output.csv", index=False)
```
