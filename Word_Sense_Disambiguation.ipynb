```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
```

```python
from nltk import pos_tag
```

```python
def get_wordnet_pos(treebank_tag):
    if treebank_tag.startswith('J'):
        return wn.ADJ
    elif treebank_tag.startswith('V'):
        return wn.VERB
    elif treebank_tag.startswith('N'):
        return wn.NOUN
    elif treebank_tag.startswith('R'):
        return wn.ADV
    else:
        return None
```

```python
import nltk
nltk.download('punkt_tab')

from nltk.wsd import lesk
from nltk.tokenize import word_tokenize
from nltk.corpus import wordnet as wn

def advanced_wsd(sentence, word):
    tokens = word_tokenize(sentence.lower())
    tokens = [w for w in tokens if w not in stop_words]

    tagged = pos_tag(tokens)

    word_pos = None
    for w, tag in tagged:
        if w == word.lower():
            word_pos = get_wordnet_pos(tag)
            break

    best_sense = None
    max_score = 0

    for sense in wn.synsets(word, pos=word_pos):
        signature = set(word_tokenize(sense.definition().lower()))

        for ex in sense.examples():
            signature.update(word_tokenize(ex.lower()))

        score = len(set(tokens).intersection(signature))

        if score > max_score:
            max_score = score
            best_sense = sense

    print("\nSentence:", sentence)
    print("Word:", word)

    if best_sense:
        print("Best Sense:", best_sense.name())
        print("Definition:", best_sense.definition())
    else:
        print("No sense found")
```

```python
test_cases = [
    ("He sat on the bank of the river", "bank"),
    ("She deposited money in the bank", "bank"),
    ("The crane is flying in the sky", "crane"),
    ("He used a crane to lift heavy objects", "crane"),
]

for sent, word in test_cases:
    advanced_wsd(sent, word)
```
