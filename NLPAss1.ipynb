```python
import nltk
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('omw-1.4')

from nltk.tokenize import word_tokenize, TreebankWordTokenizer, TweetTokenizer, MWETokenizer
from nltk.stem import PorterStemmer, SnowballStemmer
from nltk.stem import WordNetLemmatizer
import string
```

```python
text = "NLTK is great! I love learning NLP, especially tokenization & stemming. #AI"
```

```python
whitespace_tokens = text.split()
print("Whitespace:", whitespace_tokens)
```

```python
punct_tokens = [word for word in text.split() if word not in string.punctuation]
print("Punctuation-based:", punct_tokens)
```

```python
treebank = TreebankWordTokenizer()
print("Treebank:", treebank.tokenize(text))
```

```python
tweet = TweetTokenizer()
print("Tweet:", tweet.tokenize(text))
```

```python
mwe = MWETokenizer([('machine', 'learning'), ('natural', 'language', 'processing')], separator='_')
mwe_text = "I study machine learning and natural language processing"
print("MWE:", mwe.tokenize(word_tokenize(mwe_text)))
```

```python
porter = PorterStemmer()
print("Porter Stemmer:", [porter.stem(word) for word in word_tokenize(text)])
```

```python
snowball = SnowballStemmer("english")
print("Snowball Stemmer:", [snowball.stem(word) for word in word_tokenize(text)])
```

```python
lemmatizer = WordNetLemmatizer()
print("Lemmatization:", [lemmatizer.lemmatize(word) for word in word_tokenize(text)])
```
